# Problem 1

In binary classification, the distance $d$ of a point $\mathbf{x}$ to a hyperplane $\mathbf{w}$ is computed by the length of the projection of $\mathbf{x}$ onto $\mathbf{w}$, minus the distance $r$ to the origin:
$$
d =  \mathbf{x} \cdot \frac{\mathbf{w}}{\| \mathbf{w} \| }  - r
$$

Show how the term $r$ can be omitted by adding an extra dimension to $\mathbf{w}$ and $\mathbf{x}$, resulting in vectors $\mathbf{w}'$ and $\mathbf{x}'$, s.t. the scaled distance $d\ ||\mathbf{w}||$ satisfies $$d\ ||\mathbf{w}|| = \mathbf{w}' \cdot \mathbf{x}' $$



YOUR ANSWER HERE

$$
d =  \mathbf{x} \cdot \frac{\mathbf{w}}{\| \mathbf{w} \| }  - r \ | *\|\mathbf{w}\| \\
d\|\mathbf{w}\| = \mathbf{x} \mathbf{w} - r\|\mathbf{w}\| \ ^{(eq.1)}
$$

r is the threshold <br>
folding the threshold into the vector $\mathbf{w}$ <br>
add an extra component to each data point $\mathbf{x}$ and extra component to the normal $\mathbf{w}$ <br>
if $\mathbf{w} = (w_{1}, ... , w_{n})$ and $\mathbf{x} = (x_{1}, ... , x_{n})$ <br>
then: $\mathbf{w'} = (w_{1}, ... , w_{n}, r\|\mathbf{w}\|)$ and $\mathbf{x'} = (x_{1}, ... , x_{n}, -1)$ <br>
then $\mathbf{w'}\mathbf{x'} = (w_{1}, ... , w_{n}, r\|\mathbf{w}\|) \cdot (x_{1}, ... , x_{n}, -1) =^{(eq.1)} \\ \mathbf{w}\mathbf{x} -r\|\mathbf{w}\| = d\|\mathbf{w}\|$

# Problem 2

Given the definition of $d$ above, show that scaling the vector $\mathbf{w}' = \beta \mathbf{w}$ for $\beta \in \mathbb{R}^+$ does not affect the distance $d$.


YOUR ANSWER HERE
$$
\mathbf{x} \cdot \frac{\mathbf{w'}}{\| \mathbf{w'} \| }  - r = \mathbf{x} \cdot \frac{\beta \mathbf{w}}{\| \beta \mathbf{w} \| }  - r = \mathbf{x} \cdot \frac{\beta \mathbf{w}}{\beta \|\mathbf{w} \| }  - r = d
$$

# Problem 3

An alternative specification of the distance to a hyperplane is by using the difference of point $\mathbf{x}$ to a point $\mathbf{a}$ on the hyperplane:

$$ d' = (\mathbf{x}-\mathbf{a}) \cdot \frac{\mathbf{w}}{||\mathbf{w}||} $$
Show that $d' = d $ using the definition of $d =  \mathbf{x} \cdot \frac{\mathbf{w}}{\| \mathbf{w} \| } - r$

YOUR ANSWER HERE     
$$
d' = (x-a)\frac{\mathbf{w}}{\|\mathbf{w}\|}\\
d' = \mathbf{x} \cdot \frac{\mathbf{w}}{\| \mathbf{w} \| } - a \frac{\mathbf{w}}{\|\mathbf{w}\|}, \ a \frac{\mathbf{w}}{\|\mathbf{w}\|} = r\\
{\displaystyle \Rightarrow } \ d' = d
$$
