## Problem 1
Prepare data for multiclass classification from the iris dataset. It consists of 4-dimensional feature vectors for 50 data for each of three classes (0 = iris setosa, 1 = iris versicolor, 2 = iris virginica). Prepare a 120/30 split into train and test by holding out each fifth datapoint for testing.

```python
def read_iris(filename):
    """Reads the iris file into memory.

    Args:
        filename: string. The filename

    Returns:
        list of lists (one for each data point). Inner lists contain a numpy array (the features)
            and an integer (iris type).
    """
    # Your code here
    df = pd.read_csv(filename,header=None)

    df[4] = df[4].replace(['Iris-setosa'],0)
    df[4] = df[4].replace(['Iris-versicolor'],1)
    df[4] = df[4].replace(['Iris-virginica'],2)

    return df

def split_data(data):
    """Split the data into training and test sets.

    Args:
        data: Output format of read_iris()

    Returns:
        Tuple containing training set and test set (each in the input format)
    """
    # Your code here
    X = []
    y = []

    for index, rows in data.iterrows():
    # Create list for the current row
      if index%4==0:
        my_list =[rows[0], rows[1], rows[2], rows[3], rows[4]]
        y.append(my_list)
      else:
        my_list =[rows[0], rows[1], rows[2], rows[3], rows[4]]
        X.append(my_list)

    return X,y
```

## Problem 2
Implement a perceptron algorithm for multiclass classification using a block feature vector construction and apply it to the iris dataset.

```python
class MCP:
    """Multiclass Perceptron"""

    def __init__(self, features=4, labels=3, epochs=250):
        self.f = features
        self.l = labels
        self.N = epochs
        self.weights = None

    def initialize_weights(self):
        """Set self.weights to a numpy array. May contain zeros, or random numbers.
        """
        # Your code here
        #self.weights = np.random.rand(4)
        self.weights = {c: np.array([random.random() for _ in range(self.f)]) for c in range(self.l)}
        #self.weights = {c: np.array([0 for _ in range(self.f)]) for c in range(self.l)}
        return self

    def train(self, train):
        """Train the MCP on training data.

        Args:
            train: Iris data format. The training data
        """
        # Your code here.
        for _ in range(self.N):
            for feature_vect, category in zip([i[:4] for i in train],[i[4] for i in train]):
                print('feature dict', feature_vect)
                category = category.astype(int)
                print(category)


                # Initialize arg_max value, predicted class.
                arg_max, predicted_class = 0, 0

                # Multi-Class Decision Rule:
                for c in range(self.l):
                    print('weights',self.weights)
                    current_activation = np.dot(feature_vect, self.weights[c])
                    print('current activation',current_activation)
                    if current_activation >= arg_max:
                        arg_max, predicted_class = current_activation, c

                # Update Rule:
                print('predicted class',predicted_class)
                if not (category == predicted_class):
                    self.weights[category] = self.weights[category]+ feature_vect
                    self.weights[predicted_class] = self.weights[predicted_class]- feature_vect
                print('final weights', self.weights)

    def predict(self, feature_dict):
        """
        Categorize a brand-new, unseen data point based on the existing collected data.
        :param  feature_dictionary  Dictionary of the same form as the training feature data.
        :return                     Return the predicted category for the data point.
        """


        # Initialize arg_max value, predicted class.
        arg_max, predicted_class = 0, 0

        # Multi-Class Decision Rule:
        for c in range(self.l):
            current_activation = np.dot(feature_dict, self.weights[c])
            if current_activation >= arg_max:
                arg_max, predicted_class = current_activation, c

        return predicted_class

    def test(self, test):
        """Test the MCP on test data.

        Args:
            test: Iris data format. The test data

        Returns:
            float -- Accuracy on the test set
        """
        # Your code here
        correct, incorrect = 0, 0
        for feature_vect, category in zip([i[:4] for i in test],[i[4] for i in test]):
            actual_class = category
            predicted_class = self.predict(feature_vect)

            if actual_class == predicted_class:
                correct += 1
            else:
                incorrect += 1

        print("ACCURACY:")
        print("Model Accuracy:", (correct * 1.0) / ((correct + incorrect) * 1.0))
```
